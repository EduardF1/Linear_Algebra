```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# import picture
pic = Image.open('Einstein_tongue.jpg')

# show the picture
plt.imshow(pic)
plt.show()

# convert image to floating-point precision (the image is initially loaded with int values)
pic = np.array(pic)
plt.imshow(pic,cmap='gray')
plt.show()

# SVD (Singular value decomposition)
# SVD goal (purpose): deconstruct a matrix. The image shown is a matrix representation of the
# image through the numeric values of x-0-y (these values represent the pixel intensity).
# SVD : initial_matrix --decompose--> 3 result matrices U,S (Sigma),V. The result matrices
# describe spatial patterns.
# do the SVD
U, S, V = np.linalg.svd(pic)

# plot the spectrum/scree plot (spectrum = chart/graph showing the intensity of light being emitted
# over a range of energies)
plt.plot(S, '')
plt.xlim([0,100])
plt.xlabel('Component number')
plt.ylabel('Singular value (\Sigma)')
plt.show()

# reconstruct the image based on some components

# List the components used for reconstruction (first 5)
comps = np.arange(10,60)

# Reconstruct the low-rank version of the picture
reconPic = U[:,comps]@np.diag(S[comps])@V[comps,:]

#show the original and reconstructed pictures for comparison
plt.figure(figsize=(5,10))
plt.subplot(1,2,1)
plt.imshow(pic,cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(reconPic, cmap='gray')
plt.title('Components %s-%s' %(comps[0], comps[-1]))
plt.axis('off')

plt.show()
```
