```python
from matplotlib import pyplot as plt

import numpy as np

# ignore warnings
np.warnings.filterwarnings('ignore', category=np.VisibleDeprecationWarning)

# matrix S
S = np.array([[3, 0],
              [5, 2],
              [1, 2]])

# vector v
v = np.array([[-3], [1], [5]])  # vector in plane
# v = np.array([ [1], [7], [3] ])   # vector not in plane

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# draw plane corresponding to the column space
xx, yy = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10))
# cross product of the vectors contained within S
cp = np.cross(S[:, 0], S[:, 1])
# normal vector, perpendicular to the plane
z1 = (-cp[0] * xx - cp[1] * yy) / cp[2]
ax.plot_surface(xx, yy, z1, alpha=.2)

# plot the contained vectors of matrix S
ax.plot([0, S[0, 0]], [0, S[1, 0]], [0, S[2, 0]], 'k')
ax.plot([0, S[0, 1]], [0, S[1, 1]], [0, S[2, 1]], 'k')

# plot vector v
ax.plot([0, v[0]], [0, v[1]], [0, v[2]], 'r')

ax.view_init(elev=150, azim=0)
plt.show()

# Note: vector v is the red one which is inside the plane, the dark blue ones are also in the plane
```
