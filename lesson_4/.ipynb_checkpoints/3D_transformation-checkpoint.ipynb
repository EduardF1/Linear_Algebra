```python
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np
import math

v = np.array([ -3.3, 8, 0.25 ]) #initial vector


th = 5*np.pi/24 #rotation angle
Az = np.array([ [math.cos(th),-math.sin(th),0], [math.sin(th),math.cos(th),0],[0,0,1] ]) #z-rotation matrix
Ay = np.array([ [math.cos(th),0,math.sin(th)], [0,1,0],[-math.sin(th),0,math.cos(th)] ]) #y-rotation matrix
Ax = np.array([ [1,0,0], [0,math.cos(th),-math.sin(th)],[0,math.sin(th),math.cos(th)] ]) #x-rotation matrix


wz = Az@np.matrix.transpose(v) #z-rotated vector
wy = Ay@np.matrix.transpose(v) #y-rotated vector
wx = Ax@np.matrix.transpose(v) #x-rotated vector


# initial plot for the 3 vectors
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

plt.plot([0,v[0]],[0,v[1]],[0,v[2]],label='v')
plt.plot([0,wz[0]],[0,wz[1]],[0,wz[2]],label='AvZ')
plt.plot([0,wx[0]],[0,wx[1]],[0,wx[2]],label='AvX')
plt.plot([0,wy[0]],[0,wy[1]],[0,wy[2]],label='AvY')
plt.legend()
plt.grid()
```
