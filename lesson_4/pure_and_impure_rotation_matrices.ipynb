```python
import numpy as np
import matplotlib.pyplot as plt

Theta = []
magnitudes = []
magnitudes2 = []
v = np.array([3, -2])

for x in range(0, 60):
    thetas = x * 2 * np.pi / 60
    Theta.append(thetas)

    A1 = np.array([[np.cos(thetas), -np.sin(thetas)],
                   [np.sin(thetas), np.cos(thetas)]])
    A2 = np.array([[2 * np.cos(thetas), -np.sin(thetas)],
                   [np.sin(thetas), np.cos(thetas)]])

    vector1 = np.matmul(A1, v)
    vector2 = np.matmul(A2, v)
    magnitudes.append(np.linalg.norm(vector1))
    magnitudes2.append(np.linalg.norm(vector2))

    plt.figure(1, figsize=(7, 7))
    plt.axis('equal')
    plt.plot([0, vector1[0]], [0, vector1[1]], 'k')
    plt.legend(['pure', 'impure'])
    plt.plot([0, vector2[0]], [0, vector2[1]], linestyle='dashed')
    plt.grid(b=True, color='#D8D8D8')

plt.figure(2)
plt.figure(2, figsize=(7, 7))
plt.axis('equal')

plt.plot(Theta, magnitudes)
plt.plot(Theta, magnitudes2, 'r')
plt.legend(["pure", 'impure'])
```
