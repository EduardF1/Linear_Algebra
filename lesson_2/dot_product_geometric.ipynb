```python
import matplotlib.pyplot as plt
import numpy as np

# two vectors
v1 = np.array([2, 4, -3])
v2 = np.array([0,-3,-3])

# compute the angle (radians) between the two vectors
ang = np.arccos( np.dot(v1,v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))

# draw the vectors
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.plot([0, v1[0]], [0,v1[1]], [0,v1[2]],'b')
ax.plot([0, v2[0]], [0,v2[1]], [0,v2[2]],'r')

plt.axis((-6, 6, -6, 6))
plt.title('Angle between vectors: %s radians.' %ang)
plt.show()

## equivalence of algebraic and geometric dot product formulas

# two vectors
v1 = np.array([2,4,-3])
v2 = np.array([0, -3, -3])

# algebraic
dp_a = np.dot(v1, v2)

# geometric
dp_g = np.linalg.norm(v1) * np.linalg.norm(v2)*np.cos(ang)

# print dot product
print(dp_a)
print(dp_a)
```
